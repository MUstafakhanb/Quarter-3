```python
print("01_chaotic_counting")

import random

DONE_LIKELIHOOD = 0.2

def done():
    """Return True with probability DONE_LIKELIHOOD."""
    return random.random() < DONE_LIKELIHOOD

def chaotic_counting():
    """Counts from 1 to 10, but stops early if done() is True."""
    for i in range(1, 11):
        print(i)
        if done():
            break

def main():
    print("I am going to count until 10 or until I feel like stopping, whichever comes first.\n")
    chaotic_counting()
    print("\nI'm done counting.")

main()
```
