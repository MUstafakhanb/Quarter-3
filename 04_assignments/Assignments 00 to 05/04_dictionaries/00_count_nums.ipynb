```python
print("00_count_nums")

def main():
    count_nums = {}

    while True:
        user_input = input("Enter a number (or press Enter to stop):")
        if user_input == "":
            break
        number = int(user_input)
        if number in count_nums:
            count_nums[number] += 1
        else:
            count_nums[number] = 1
    print("\nNumber occurrences:")
    for num, count in count_nums.items():
        print(f"{num} appears {count} times.")

if __name__ == '__main__':
    main()
```
